```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
# Load sales data and print first 5 items
# SALES = pd.read_csv("../data/items_purchased.csv")
SALES = pd.read_csv("~/Desktop/items_purchased.csv")
SALES.head()
```

```python
# List 300 most popular products
# Very curious to know what product_id 4061 is. Chips? 
report = SALES.groupby('product_id').product_id.count().reset_index(name ='count')
report = report.sort_values(by='count', ascending=False)

print("300 Most Popular Products")
report.head(n=300)

#report.plot.bar();
```

```python
# Most popular products at each kiosk
report = SALES.groupby(['kiosk_id', 'product_id']).size().reset_index(name='count')
report = report.sort_values(by='count', ascending=False)

print("Most Popular Products By Kiosk")
report.head(n=100)
```

```python
# Most popular product categories
report = SALES.groupby('fc_number').product_id.count().reset_index(name ='count')
report = report.sort_values(by='count', ascending=False)

print("Most Popular Product Categories")
report.head(n=46)
```

```python
# Most lucrative kiosks
report = SALES.groupby('kiosk_id').product_id.count().reset_index(name ='count')
report = report.sort_values(by='count', ascending=False)

print("50 Most Lucrative Kiosks")
report.head(n=50)
```
